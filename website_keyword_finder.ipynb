```python
import time

try:
    from colorama import Fore,Style,init
    init(autoreset=True)
except:
    class X:
        def __getattr__(self,n): return ""
    Fore=Style=X()

website="""Welcome to TechWorld!

Artificial Intelligence and Machine Learning are transforming the future.
Python is widely used for AI, Data Science, Web Development and Cloud Computing.
ChatGPT helps developers write code faster.
Cyber Security protects websites and applications.
"""

def naive_search(t,p):
    m,n=len(p),len(t);comp=0;pos=[]
    for i in range(n-m+1):
        j=0
        while j<m:
            comp+=1
            if t[i+j]!=p[j]: break
            j+=1
        if j==m: pos.append(i)
    return pos,comp

def lps(p):
    l=[0]*len(p);i=1;ln=0
    while i<len(p):
        if p[i]==p[ln]:
            ln+=1;l[i]=ln;i+=1
        elif ln: ln=l[ln-1]
        else: i+=1
    return l

def kmp_search(t,p):
    L=lps(p);i=j=0;comp=0;pos=[]
    while i<len(t):
        comp+=1
        if t[i]==p[j]:
            i+=1;j+=1
        if j==len(p):
            pos.append(i-j);j=L[j-1]
        elif i<len(t) and t[i]!=p[j]:
            j=L[j-1] if j else 0
            if j==0 and t[i]!=p[0]: i+=1
    return pos,comp

def rk_search(t,p,q=101):
    d=256;m=len(p);n=len(t)
    h=pow(d,m-1,q);ph=th=0;comp=0;pos=[]
    for i in range(m):
        ph=(d*ph+ord(p[i]))%q
        th=(d*th+ord(t[i]))%q
    for s in range(n-m+1):
        if ph==th:
            ok=True
            for k in range(m):
                comp+=1
                if t[s+k]!=p[k]:
                    ok=False;break
            if ok: pos.append(s)
        if s<n-m:
            th=(d*(th-ord(t[s])*h)+ord(t[s+m]))%q
    return pos,comp

def compare(keyword):
    for name,func in [("Naive",naive_search),("KMP",kmp_search),("Rabin-Karp",rk_search)]:
        st=time.perf_counter()
        pos,comp=func(website.lower(),keyword.lower())
        tm=(time.perf_counter()-st)*1000
        print(Fore.CYAN+f"\n{name}")
        print("Found:", "Yes" if pos else "No")
        print("Positions:",pos)
        print("Comparisons:",comp)
        print(f"Time: {tm:.6f} ms")

while True:
    print(Fore.GREEN+"\n===== WEBSITE KEYWORD FINDER =====")
    print("1. Display Website")
    print("2. Search Keyword")
    print("3. Exit")
    ch=input("Choice: ")
    if ch=="1":
        print(Fore.YELLOW+website)
    elif ch=="2":
        kw=input("Enter keyword: ")
        compare(kw)
    elif ch=="3":
        break
    else:
        print("Invalid choice")
```
